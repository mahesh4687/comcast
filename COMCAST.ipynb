# PROVIDE STATE WISE STATUS OF COMPLAINTS IN A STACKED BAR CHART. USE THE CATEGORIZED VARIABLE FROM Q3. PROVIDE INSIGHTS ON

```python
comcast_df['State'] = comcast_df['State'].str.title()
```

```python
state_wise_status=comcast_df.groupby(['State','updated_status']).size().unstack().fillna(0)
```

```python
state_wise_status
```

```python
state_wise_status.sort_values('Closed',axis=0,ascending=True).plot(kind='barh',figsize=(10,8), stacked=True, title= "STACKED BAR CHART OF STATE WISE COMPLAINTS")
```
